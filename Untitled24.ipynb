```python
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print("ЧАСТЬ 1: ЗАГРУЗКА УЛИК И ПЕРВЫЙ ОСМОТР")
print("Первые 7 строк таблицы:")
print(df.head(7))
print("\nТехническая сводка о данных:")
print(df.info())
print("\nСтатистическая сводка:")
print(df.describe())

print("\nЧАСТЬ 2: ОЧИСТКА ДАННЫХ И ПОДГОТОВКА К АНАЛИЗУ")
median_age = df['Age'].median()
print(f"Медианный возраст: {median_age:.2f} лет")
df['Age'].fillna(median_age, inplace=True)
print(f"Пропуски в Age после заполнения: {df['Age'].isnull().sum()}")

def get_age_group(age):
    if age < 18:
        return "Ребенок"
    elif age <= 65:
        return "Взрослый"
    else:
        return "Пожилой"

df['AgeGroup'] = df['Age'].apply(get_age_group)
print("\nПроверка AgeGroup:")
print(df[['Age', 'AgeGroup']].head())

print("\nЧАСТЬ 3: ПОИСК ЗАКОНОМЕРНОСТЕЙ")
print("\n3.1 Анализ выживаемости по полу:")
survival_by_sex = df.groupby('Sex')['Survived'].mean() * 100
print(survival_by_sex)

print("\nАнализ выживаемости по классу:")
survival_by_class = df.groupby('Pclass')['Survived'].mean() * 100
print(survival_by_class)

print("\nСводная таблица по полу и классу:")
pivot_table = df.pivot_table(values='Survived', index='Pclass', columns='Sex', aggfunc='mean') * 100
print(pivot_table)

print("\n3.2 Фильтрация и сортировка:")
minors_3rd_class = df[(df['Age'] < 18) & (df['Pclass'] == 3) & (df['Survived'] == 1)]
minors_sorted = minors_3rd_class[['Name', 'Age']].sort_values('Age', ascending=False)
print("Несовершеннолетние в 3 классе (выжившие):")
print(minors_sorted)

oldest_dead_man = df[(df['Sex'] == 'male') & (df['Survived'] == 0)].sort_values('Age', ascending=False).head(1)
print("\nСамый пожилой погибший мужчина:")
print(oldest_dead_man[['Name', 'Age', 'Pclass']].values)

print("\n3.3 Анализ влияния размера семьи:")
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df['IsAlone'] = (df['FamilySize'] == 1).astype(int)
survival_by_alone = df.groupby('IsAlone')['Survived'].mean() * 100
print("Выживаемость по одиночеству:")
print(survival_by_alone)

print("\nЧАСТЬ 4: ВЫВОДЫ")
print("\nВлияние пола на выживаемость:")
print(f"Женщины: {survival_by_sex['female']:.1f}%")
print(f"Мужчины: {survival_by_sex['male']:.1f}%")

print("\nВлияние класса на выживаемость:")
for pclass, rate in survival_by_class.items():
    print(f"Класс {pclass}: {rate:.1f}%")

print("\nАнализ правила 'женщины и дети в первую очередь':")
survival_by_age_group = df.groupby('AgeGroup')['Survived'].mean() * 100
print("\nВыживаемость по возрастным группам:")
print(survival_by_age_group)

women_children = df[(df['Sex'] == 'female') | (df['AgeGroup'] == 'Ребенок')]
others = df[(df['Sex'] == 'male') & (df['AgeGroup'] != 'Ребенок')]
wc_survival = women_children['Survived'].mean() * 100
others_survival = others['Survived'].mean() * 100

print(f"\nВыживаемость женщин и детей: {wc_survival:.1f}%")
print(f"Выживаемость взрослых мужчин: {others_survival:.1f}%")

print("\nДополнительная статистика:")
print(f"Всего пассажиров: {len(df)}")
print(f"Выжило: {df['Survived'].sum()} ({df['Survived'].mean()*100:.1f}%)")
print(f"Средний возраст: {df['Age'].mean():.1f} лет")
print(f"Медианная цена билета: {df['Fare'].median():.2f}")
```
